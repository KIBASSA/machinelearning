```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.applications import ResNet50, MobileNet,InceptionV3
from keras.applications.resnet50 import preprocess_input, decode_predictions
from keras.preprocessing.image import load_img,img_to_array
```

```python
model = ResNet50(weights="imagenet")
```

```python
#load image 
image_path  = "images/car.jpg"
img_pil = load_img(image_path, target_size=(224,224))
#convert to array
img_array = img_to_array(img_pil)
image_batch = np.expand_dims(img_array,axis=0)
```

```python
image_preprocess = preprocess_input(image_batch)

image_predicted = model.predict(image_preprocess)

print('Predicted:', decode_predictions(image_predicted, top=3)[0])
```

```python
##### mobile net
model_mobileNet = MobileNet(weights="imagenet")
```

```python
image_mobilenet_pil = load_img(image_path, target_size=(224,224))
plt.imshow(image_mobilenet_pil)
```

```python
image_mobilenet_array = img_to_array(image_mobilenet_pil)
plt.imshow(np.uint8(image_mobilenet_array))
```

```python
image_mobilenet_batch = np.expand_dims(image_mobilenet_array,axis=0)
```

```python
image_mobilenet_preprocess = preprocess_input(image_mobilenet_batch)

image_mobilenet_predict = model_mobileNet.predict(image_mobilenet_preprocess)

label_mobilenet = decode_predictions(image_mobilenet_predict)
```

```python
print(label_mobilenet)
```

```python
#InceptionV3

model_InceptionV3 = InceptionV3(weights="imagenet")
```
