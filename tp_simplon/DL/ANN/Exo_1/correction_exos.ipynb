## Pre-processing

### Importation des libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_excel('data/Churn_Modelling.xlsx')
X = dataset.iloc[:, 3:13].values
y = dataset.iloc[:, 13].values
```

```python
# Encoding categorical data (geography et gender)
from sklearn.preprocessing import LabelEncoder, OneHotEncoder

labelencoder_X_1 = LabelEncoder()
X[:, 1] = labelencoder_X_1.fit_transform(X[:, 1])

labelencoder_X_2 = LabelEncoder()
X[:, 2] = labelencoder_X_2.fit_transform(X[:, 2])


onehotencoder = OneHotEncoder(categorical_features = [1])
X = onehotencoder.fit_transform(X).toarray()
X = X[:, 1:]
```

* Splitting the dataset into the Training set and Test set
* test_size = % à mettre dans le test set
* random_state = Pour avoir les mêmes données d'un ordi à l'autre

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

## Part 2 - Now let's make the ANN!

```python
# Importing the Keras libraries and packages
import keras
from keras.models import Sequential
#Permet de créer les couches de neuronnes
from keras.layers import Dense
#Permet de faire des tests en supprimant aléatoirement des neuronnes
from keras.layers import Dropout
```

```python
# Initialising the ANN
classifier = Sequential()
```

* Adding the input layer and the first hidden layer
* units = corresponds au nombre de neurones sur une couche
* kernel_initializer = permet de définir la manière dont les poids seront initialisés
* activation = permet de définir la fonction d'activation d'un neurone (sigmoid, Relu, tanh, etc.)
* relu = fonction redresseur
* input_dim = c'est le nombre des features de l'input 

```python
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
# rate = 0.1 : Le neuronne à 10% de chance d'être désactivé
classifier.add(Dropout(rate=0.1))
```

```python
# Adding the second hidden layer
classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
classifier.add(Dropout(rate=0.1))
```

### Adding the output layer
* On veut une probabilité en sortie, on utilise une fonction sigmoid
* Si on a plus de 2 catégories, on utilise activation = 'softmax'

```python
classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
```

### Compiling the ANN
* On utilise une fonction de cout logarithmique 'loss = binary_crossentropy' car on est dans le cas d'une reg logistique
* loss = 'categorical_crossentropy si on avait plus d'une variable binaire
* metrics = sont des fonctions utilisées pour juger la performance du modèle 

```python
classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

### Fitting the ANN to the Training set
* batch_size = nombre de lots d'observations
* epochs = nombre d'itérations (forward + Backward)

```python
classifier.fit(X_train, y_train, batch_size = 10, epochs = 100)
```

## Part 3 - Making predictions and evaluating the model

### Juger le modèle
 algorithme de validation croisé (permet de splitter le training set en plein 
 de morceau et d'isoler une valeur pour la tester)

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score

def build_classifier():
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
#K-Fold cross validation
classifier = KerasClassifier(build_fn=build_classifier, batch_size = 10, epochs = 100)
# , n_jobs=-1 à la fin :
precision = cross_val_score(estimator=classifier, X=X_train, y=y_train, cv=10)

#apprécier la moyenne des différents tests et l'écart type
moyenne = precision.mean()
ecart_type = precision.std()
```

# Grid search

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import GridSearchCV
```

```python
#on spécifie la fonction qui renvoie le classifier avec en parametre 'optimizer'
def build_classifier(optimizer):
    classifier = Sequential()
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu', input_dim = 11))
    classifier.add(Dense(units = 6, kernel_initializer = 'uniform', activation = 'relu'))
    classifier.add(Dense(units = 1, kernel_initializer = 'uniform', activation = 'sigmoid'))
    classifier.compile(optimizer = optimizer, loss = 'binary_crossentropy', metrics = ['accuracy'])
    return classifier
```

```python
#K-Fold cross validation
classifier = KerasClassifier(build_fn=build_classifier)
parameters = {"batch_size" = [25, 32],
              "nb_epoch" = [100, 500],
              "optimizer" = {"adam", "rmsprop"}}
grid_search = GridSearchCV(estimator = classifier, param_grid = parameters, scoring = 'accuracy', cv = 10)
grid_search = grid_search.fit(X_train, y_train)
```

```python
best_parameters = grid_search.best_params_
best_accuracy = grid_search.best_score_
```
