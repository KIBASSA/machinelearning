On remplace dans la colonne review_score à 1 si la note est supérieure à 3 et -1, si le contraire 

```r
binaryratings <- customer_product_rating
for (i in 1:nrow(customer_product_rating)){
 if (customer_product_rating[i,3] > 3){
   binaryratings[i,3] <- 1
 }
 else{
   binaryratings[i,3] <- -1
 }
}
```

Vue que l'opération prend beaucoup de place mémoire, on sélection juste une partie 

```r
binaryratings_compressed <- binaryratings[1:100,]
binaryratings_compressed
```

On utilise **dcast** pour créer une matrice à partir d'un format long de binaryratings_compressed à un format large avec comme valeur interne **review_score**

* En première ligne, c'est la liste des clients 
* En première colonne, c'est la liste des produits
* Et chaque cellule correspond à un client qui a acheté un produit (1) ou non (0). 

```r
binaryratings2 <- dcast(binaryratings_compressed, product_id~customer_id, value.var = "review_score", fun.aggregate = 
                        function(x) {
                            paste(x, collapse = "; ")
                        } , fill = "0")

for (i in 1:ncol(binaryratings2)){
  binaryratings2[which(is.na(binaryratings2[,i]) == TRUE),i] <- 0
}
binaryratings2 = binaryratings2[,-1] #remove product_id col. Rows are product_ids, cols are customer_ids
```

### Creation du profile client

Avant de créer le profile client, nous transformons les valeurs de la matrice binaryratings2 en numérique. 
Car c'était en string à cause de l'opération "**paste(x, collapse = "; ")**"

```r
for (c in 1:ncol(binaryratings2)){
    for (r in 1:nrow(binaryratings2)){
        if (binaryratings2[r,c] == "0" || binaryratings2[r,c] == "1" || binaryratings2[r,c] == "-1")
            {
             binaryratings2[r,c] <- as.numeric(binaryratings2[r,c])      
        }
        else
            {
             arr <- strsplit(binaryratings2[r,c], ";")[[1]]
             binaryratings2[r,c] <- 0
             for (f in arr)
                 {
                  binaryratings2[r,c] <- as.numeric(binaryratings2[r,c]) + as.numeric(f)       
                 }
             binaryratings2[r,c] <- as.numeric(binaryratings2[r,c])
        }
    }
}
```

Now we can calculate the dot product of the category matrix and the ratings matrix and obtain the user profiles.

```r
#Calculate dot product for User Profiles
user_profile = matrix(0,length(categories_list),nrow(Customers))
for (c in 1:ncol(binaryratings2)){
  for (i in 1:ncol(category_matrix2)){#category_matrix3
      #print(category_matrix2[,i])
      user_profile[i,c] <- sum((category_matrix2[,i]) * (as.numeric(binaryratings2[,c])))
  }
}
```

Les valeurs obtenues sont converti en 1 si c'est positif et en 0 si c'est négatif

```r
#Convert to Binary scale
for (i in 1:nrow(user_profile)){
  if (user_profile[i] < 0){
    user_profile[i] <- 0
  }
  else {
    user_profile[i] <- 1
  }
}
```
