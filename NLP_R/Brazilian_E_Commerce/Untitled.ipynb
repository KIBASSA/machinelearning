```r
library(tidytext)
#Bigram
book_bigram <- data_frame(line = 1:nrow(book), text = book$text)  %>%
  unnest_tokens(bigram, text, token = "ngrams", n = 2) %>%
  count(bigram, sort = TRUE)
#Trigram
book_trigram <- data_frame(line = 1:nrow(book), text = book$text)  %>%
  unnest_tokens(bigram, text, token = "ngrams", n = 3)%>%
  count(bigram, sort = TRUE)
#Four-gram 
book_fourgram <- data_frame(line = 1:nrow(book), text = book$text)  %>%
  unnest_tokens(bigram, text, token = "ngrams", n = 4)%>%
  count(bigram, sort = TRUE)
#et ainsi de suite ! 
```

NETTOYER LE TEXTE

```r
library(tidytext)
#Bigram
book_bigram_clean <- book_bigram %>%
  separate(bigram, c("word1", "word2"), sep = " ") %>%
  filter(!word1 %in% stop_words$word) %>%
  filter(!word2 %in% stop_words$word) %>%
  unite(bigram, word1, word2, sep = " ")
#Trigram
book_trigram_clean <- book_trigram  %>%
  separate(bigram, c("word1", "word2", "word3"), sep = " ") %>%
  filter(!word1 %in% stop_words$word) %>%
  filter(!word3 %in% stop_words$word) %>%
  unite(bigram, word1, word2, word3, sep = " ")
```

### Text mining et topic modeling avec R

source : https://thinkr.fr/text-mining-et-topic-modeling-avec-r/

Cette approche sutilise plus particulirement lors de lanalyse de corpus (une collection de documents, un document est une collection de thmatiques, et une thmatique est une collection de mots), afin de dcouvrir les spcificits propres  chaque document de la collection, afin de les regrouper par similarit. Lobjectif tant de pouvoir trier, comprendre, indexer et rsumer de larges sries de documents.

Nous allons utiliser **Allocation de Dirichlet latente** : un modle probabiliste expliquant des observations  laide de groupes (non observs) de donnes similaires.

Comment a marche ? On commence par dfinir un nombre k de thmes. Ensuite, lalgorithme ira partitionner les documents, et vous rpondra de deux manires :
+ en attribuant  chaque mot une probabilit davoir t gnr par chaque thme.
+ en attribuant  chaque document une probabilit dappartenance  chaque thme.

Dans R, une fois votre algorithme tourn, vous obtiendrez une liste de rsultats, avec deux matrices  regarder :

 + beta, qui contient les probabilits de chaque mot davoir t gnr par un topic.
 + gamma, avec la probabilit de chaque document dappartenir  un topic.

Pour raliser le topic modeling, nous aurons besoin du package {topicmodels}. Et pour lexemple, commenons par un corpus de documents simple : quatre pages Wikipdia avec quatre topics trs diffrents.

 + Marcel Proust
 + Natural logarithm
 + Kreator
 + Open-source model

```r
# Si besoin 
install.packages("topicmodels")
install.packages("tidytext")
install.packages("tidyverse")
install.packages("rvest")

# Charger les packages 
library(tidytext)
library(topicmodels)
library(tidyverse)
library(rvest)
```

```r
# Crer le tibble avec les textes
## Une fonction pour scraper et crer un df 

scrape_wiki <- function(url){
  url <- read_html(url)
  df <- tibble(text = url %>% html_nodes("p") %>% html_text(), 
               name = url %>% html_nodes("h1") %>% html_text())
  return(df)
}
url_list <- c("https://en.wikipedia.org/wiki/Marcel_Proust",
              "https://en.wikipedia.org/wiki/Natural_logarithm",
              "https://en.wikipedia.org/wiki/Kreator", 
              "https://en.wikipedia.org/wiki/Open-source_model")
data <- map_df(.x = url_list, .f = scrape_wiki)
```

Nous voici donc avec un corpus de 4 documents. Le package {topicmodel} travaillant avec des DocumentTermMatrix, nous devons faire appel  la fonction cast_dtm, de {tidytext}. (Si vous avez besoin de vous rafrachir la mmoire sur tidytext, nous vous conseillons de relire les deux billets que nous avons cits dans lintroduction ;))

```r
# Gnrons le tibble avec le compte des mots, avant de le transformer en DTM
dtm <- data %>%
  unnest_tokens(word, text) %>%
  count(name, word, sort = TRUE) %>%
  anti_join(tidytext::stop_words) %>%
  #Tranformer en DocumentTermMatrix
  cast_dtm(name, term = word, value = n)

class(dtm)
#"DocumentTermMatrix"    "simple_triplet_matrix"
# Crer le modle
lda_model <- LDA(x = dtm, k = 4, control = list(seed = 2811))
```
